```python
# time consuming...
from unstructured.partition.pdf import partition_pdf
from langchain_text_splitters import CharacterTextSplitter

def pdf_elements(path, fname):
    return partition_pdf(
        filename=path + fname,
        extract_images_in_pdf=True,
        infer_table_structure=True,
        chunking_strategy="by_title",
        max_characters=4000,
        new_after_n_chars=3800,
        combine_text_under_n_chars=2000,
        image_output_dir_path=path,
    )


def generate_texts_n_tables(pdf_elements):
    tables = []
    texts = []
    for ele in pdf_elements:
        if "unstructured.documents.elements.Table" in str(type(ele)):
            tables.append(str(ele))
        elif "unstructured.documents.elements.CompositeElement" in str(type(ele)):
            texts.append(str(ele))
    return texts, tables


# Change files directory and file names for development purpose onl!
fpath = "/content/rag/"
fname = "x.pdf"


pdf_elements = pdf_elements(fpath, fname)
texts, tables = generate_texts_n_tables(pdf_elements)

# Optional:specific token size for texts
text_splitter = CharacterTextSplitter.from_tiktoken_encoder(
    chunk_size=4000, chunk_overlap=0
)
joined_texts = " ".join(texts)
texts_token = text_splitter.split_text(joined_texts)
# texts,tables
```

```python
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI

def generate_text_summaries(texts, tables,summarize_texts=False):

    # prompt_text stole from medium haha : < )
    prompt_text = """You are an assistant tasked with summarizing tables and text for retrieval. \
    These summaries will be embedded and used to retrieve the raw text or table elements. \
    Give a concise summary of the table or text that is well optimized for retrieval. Table or text: {element} """
    prompt = ChatPromptTemplate.from_template(prompt_text)

    # using 3.5 turbo research for cheaper
    model = ChatOpenAI(temperature=0, model="gpt-3.5-turbo",api_key = API)
    summarize_chain = {"element": lambda x: x} | prompt | model | StrOutputParser()

    text_summaries = []
    table_summaries = []

    if texts and summarize_texts:
        text_summaries = summarize_chain.batch(texts, {"max_concurrency": 5})
    elif texts:
        text_summaries = texts

    if tables:
        table_summaries = summarize_chain.batch(tables, {"max_concurrency": 5})

    return text_summaries, table_summaries
```

```python
# Fetch summaries
text_summaries, table_summaries = generate_text_summaries(
    texts_token, tables, summarize_texts=True
)

text_summaries,table_summaries
```

```python
import base64
import os

from langchain_core.messages import HumanMessage


def encode_image(image_path):
    # base64 encoding
    with open(image_path, "rb") as image_file:
        return base64.b64encode(image_file.read()).decode("utf-8")


def image_summarize(img_base64, prompt):
    # costiliest part maybe reduce quality here who cares
    chat = ChatOpenAI(model="gpt-4-turbo", max_tokens=1024,api_key = API)

    msg = chat.invoke(
        [
            HumanMessage(
                content=[
                    {"type": "text", "text": prompt},
                    {
                        "type": "image_url",
                        "image_url": {"url": f"data:image/jpeg;base64,{img_base64}"},
                    },
                ]
            )
        ]
    )
    return msg.content


def generate_img_summaries(path):
    # Store base64 encoded images
    img_base64_list = []

    image_summaries = []

    prompt = """You are an assistant tasked with summarizing images for retrieval. \
    These summaries will be embedded and used to retrieve the raw image. \
    Give a concise summary of the image that is well optimized for retrieval."""

    # write bash script instead
    for img_file in sorted(os.listdir(path)):
        if img_file.endswith(".jpg"):
            img_path = os.path.join(path, img_file)
            base64_image = encode_image(img_path)
            img_base64_list.append(base64_image)
            image_summaries.append(image_summarize(base64_image, prompt))

    return img_base64_list, image_summaries


# Image summaries --> change directory of images
p = "/content/rag/figures/"
img_base64_list, image_summaries = generate_img_summaries(p)
img_base64_list,image_summaries
```

Vector Stores. Use multi_vector_retriever

```python
import uuid

from langchain.retrievers.multi_vector import MultiVectorRetriever
from langchain.storage import InMemoryStore
from langchain_community.vectorstores import Chroma
from langchain_core.documents import Document
from langchain_openai import OpenAIEmbeddings


def multi_vector_retriever(
    vectorstore, text_summaries, texts, table_summaries, tables, image_summaries, images
):
    # Initialize the storage layer
    store = InMemoryStore()
    id_key = "some_id"
    # use id_key = "doc_id"

    # Create the multi-vector retriever
    retriever = MultiVectorRetriever(
        vectorstore=vectorstore,
        docstore=store,
        id_key=id_key,
    )

    def add_docs(retriever, doc_summaries, doc_contents):
        doc_ids = [str(uuid.uuid4()) for _ in doc_contents]
        summary_docs = [
            Document(page_content=s, metadata={id_key: doc_ids[i]})
            for i, s in enumerate(doc_summaries)
        ]
        retriever.vectorstore.add_documents(summary_docs)
        retriever.docstore.mset(list(zip(doc_ids, doc_contents)))

    if text_summaries:
        add_docs(retriever, text_summaries, texts)

    if table_summaries:
        add_docs(retriever, table_summaries, tables)

    if image_summaries:
        add_docs(retriever, image_summaries, images)

    return retriever


# !!! Give APPRORIATE COLLECTION NAME
# also scope of improvement on embeddings find cheaper embedding method
vectorstore = Chroma(
    collection_name="eeg_paper", embedding_function=OpenAIEmbeddings(api_key = API)
)


retriever_multi_vector = multi_vector_retriever(
    vectorstore,
    text_summaries,
    texts,
    table_summaries,
    tables,
    image_summaries,
    img_base64_list,
)
```

```python
from langchain_core.runnables import RunnablePassthrough

# Prompt template
template = """Answer the question based only on the following context, which can include text and tables:
{context}
Question: {question}
"""
prompt = ChatPromptTemplate.from_template(template)
model = ChatOpenAI(temperature=0, model="gpt-4-turbo",api_key = API)

# RAG pipeline
chain = (
    {"context": retriever_multi_vector, "question": RunnablePassthrough()}
    | prompt
    | model
    | StrOutputParser()
)
# 6 images are all run through gpt-4-turbo and cost is boomed, find optimization here maybe
# in five files many images damn
```
